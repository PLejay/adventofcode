```python
# Read the data

def print_grid(grid):
  for row in grid:
    print(row)

# Open data file in read mode
data = open("data.txt", "r")
lines = data.read().splitlines()
print_grid(lines)
print(lines)
```

```python
# Build a grid of valid cells
grid_width = len(lines[0])
grid_height = len(lines)
max_x_index = grid_width - 1
max_y_index = grid_height - 1

# Initially filled with 0, of the same size as the data grid
grid = []
for i in range(0, grid_height):
  grid.append([0] * grid_width)
print(f"initial grid: {grid}")

# Then fill the grid with 1 to indicate valid cells

# Check whether a character is a valid symbol
def is_symbol(char: str):
  return not(char.isnumeric() or char == ".")

# Check whether a set of coordinates are accepted
def are_accepted_coordinates(x, y):
  if x >= 0 and x <= max_x_index and y >= 0 and y <= max_y_index:
    return True
  return False

# Given the coordinates of a valid symbol, set the surrounding cells to 1
def update_grid_for_valid_cells(origin):
  x, y = origin["x"], origin["y"]

for i, line in enumerate(lines):
  for j, char in enumerate(line):
    if not is_symbol(char):
      continue
    # If char is symbol, set all surrounding cells to 1
    for y in range(i - 1, i + 2):
      for x in range(j - 1, j + 2):
        if are_accepted_coordinates(x, y):
          grid[y][x] = 1


print("filled grid:")
print_grid(grid)
```

```python
# Get the sum of all valid part numbers
valid_num_sum = 0

# Check if a number spanning given coordinates is valid, using the grid built earlier
def is_valid_number(y, x_start, x_end):
  is_valid = False
  for x in range(x_start, x_end + 1):
    if grid[y][x] == 1:
      is_valid = True
  return is_valid

for i, line in enumerate(lines):
  current_num = 0
  current_num_string = ""
  current_num_start_index = -1
  current_num_end_index = -1
  for j, char in enumerate(line):
    if char.isnumeric():
      # If reaching the start of a new number, set the start index
      if len(current_num_string) == 0:
        current_num_start_index = j
      # Either way, also update the end index and the number string
      current_num_end_index = j
      current_num_string = current_num_string + char

      # If reaching the end of a number, check if the number is valid and add it to the sum if so
      if j == max_x_index or not line[j + 1].isnumeric():
        # Add the number to the sum if valid
        if is_valid_number(i, current_num_start_index, current_num_end_index):
          valid_num_sum += int(current_num_string)
          # Reset the current number
        current_num_string = ""
        current_num_start_index = -1
        current_num_end_index = -1


print(f"sum of valid part numbers: {valid_num_sum}")
```
